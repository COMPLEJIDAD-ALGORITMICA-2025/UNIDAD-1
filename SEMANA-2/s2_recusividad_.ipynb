# **Recursividad**
* La recursividad se entiende sabiendo que es la recusividad
* FUncion funcion recursiva es una funcion que se llama asi misma
* Una funcion recursiva puede ser de cálculo, te da un resultado
* Una funcion recursiva puede ser de proceso, no te regresa nada. Ordenar, imprimir, etc.
* La recursividad se entiende sabiendo que es un proceso iterativo

## **Caracteristicas de la recursividad**
1. Condicion de parada (i<=10, i<11)
2. Llamada a la misma función
3. **Contador**, este contador debe de acercarnos a la condición de parada




```python
for i in range(0, 11, 2): #nos acerca i<11
  print(i) #LLamada, el contador es i
```

* En un proceso recursivo el bloque o la instrucion que está antes de la llamada, éste se ejecuta cuando se avanza
* En un proceso recursivo el bloque o la instrucion que está después de la llamada, éste se ejecuta cuando se regresa.
*  La recusividad se ejecuta como un yoyó

```python
def imprime(n):
  if n==10:
    print(n)
    #Ya no hay llamada
    #return; #caba
  else:
    print(n)
    imprime(n+2)

imprime(0)
```

```python
def imprime(n):
  if n==0:
    print(n)
    #Ya no hay llamada
    #return; #caba
  else:
    imprime(n-2)
    print(n)

imprime(10)
```

* En un calculo recursivo el bloque o la instrucion que está antes de la llamada, éste se ejecuta cuando se avanza
* En un calculo recursivo el bloque o la instrucion que está después de la llamada, NUNCA se ejecuta.
* La recusividad se ejecuta como un yoyó, se desenrrolla y se enrrolla

```python
def factorial(n):
  if n==0 or n==1:
    print("Se proceso e: ", n)
    return 1
  else:
    print("Se proceso e: ", n)
    return n*factorial(n-1)
    #print("Se proceso e: ", n) #Esto nunc a se ejecutar'a, está depues de un return

factorial(5)
```
