```python
#Adivinar una clave de tres digitos
def FBAdivina(clave):
  intentos=0
  #hay que armar las claves
  for c in range(10):
    for d in range(10):
      for u in range(10):
        intentos+=1
        cla=str(c)+str(d)+str(u)
        if cla==str(clave):
          print("Clave encontrada")
          print("Cantidad de intentos: ", intentos)
          return

#Prueba
FBAdivina(123)
```

# BAcktraking sin poda (puro)
* Para que sea furza bruta se tendria que generar todos los numeros posibles binarios hasta de tres digitos

```python
#Dado [0,1], genere numeros binarios de 3 digitos
def generaNumerosBinarios(n, actual=""): #Lo snumero sbinarios se forman en la variable actual
  if len(actual)==n:
    print(actual)
  else:
    #Llamemos la misma funcion para probar con 0 y 1
    generaNumerosBinarios(n, actual+"0")
    generaNumerosBinarios(n, actual+"1")

generaNumerosBinarios(3)
```

```python
#Dada un alista [1,2,3], generar todas las permutaciones posible
def permutaciones(lista, actual=[]):
  if len(actual)==len(lista):
    print(actual)
    return
  else:
    for num in lista:
      if num not in actual:
        permutaciones(lista, actual+[num])

permutaciones([1,2,3])
```

Backtraking con poda

```python
#Problema de las n reynas

def es_valido(tablero, fila, columna): #Validar la solucion parcial
  #verificamos la columna
  for i in range(fila):
    if tablero[i]==columna:
      return False

  #Verificar las diagonales
  for i in range(fila):
    if abs(tablero[i]-columna)==abs(i-fila):
      return False

  return True

def nreynas(N, fila=0, tablero=[]):
  if fila==N:
    print(tablero)
    return
  for col in range(N):
    if es_valido(tablero, fila, col):
      nreynas(N, fila+1, tablero+[col]) #backtraking
  return False

nreynas(4)
```

* Hay una reyba eb la fila 0, columna 1
* Hay una reyba eb la fila 1, columna 3
* Hay una reyba eb la fila 2, columna 0
* Hay una reyba eb la fila 3, columna 2
