# Clases

```python
class CEstudiante():

  def __init__(self, _dni, _nombre, _edad):
    self.dni=_dni
    self.nombre=_nombre
    self.edad=_edad
  #Seters y geters

  def mostrar_detalle(self):
    print("Dni : ", self.dni)
    print("Nombre : ", self.dni)
    print("Edad : ", self.dni)

estudiante=CEstudiante(12154512, "Pepe ", 15)
estudiante.mostrar_detalle()

```
